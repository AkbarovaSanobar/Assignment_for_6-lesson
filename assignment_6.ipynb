```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style
import matplotlib as mpl
import matplotlib.animation as pltan
import plotly.express as px
```

```python
ann = pd.read_csv('real-commodity-price-index-food-products.csv')
ann
```

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
x = ann.Year
y = ann.Cocoa
z = ann.Coffee
v = ann.Tea
plt.style.use('seaborn-darkgrid')
plt.plot(x, y, color = "red")
plt.plot(x, z, color = 'green')
plt.plot(x, v, color = 'yellow')
plt.legend(['Cocoa', 'Coffee', 'Tea'], loc="center", bbox_to_anchor=(1.3, 0.7), shadow=True, ncol=1)

plt.xlabel('year', fontdict = font2)
plt.ylabel('Price index', fontdict = font2)
plt.xticks(rotation=90)
plt.tight_layout()
plt.grid()
plt.title("Price index of brewed drinks 1850-2015", color = 'blue', fontdict = font1, loc = 'center')
plt.figsize=(25, 20)
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
#plt.style.use('bmh')
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title('Real commodity price index of grains during 1850-2015', color='blue')
plt.ylabel('price index', color='red')
plt.xlabel('grains', color='red')

labels = ['Wheat', 'Rye', 'Rice', 'Corn', 'Barley']
plt.boxplot([ann.Wheat, ann.Rye, ann.Rice, ann.Corn, ann.Barley], labels = labels)

plt.grid()
plt.show()
```
